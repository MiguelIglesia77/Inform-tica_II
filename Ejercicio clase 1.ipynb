```python
BD= {}
class persona:
    def __init__(self, n, c, g):
        self.__nombre=n
        self.__cedula=c
        self.__genero=g
    def get_nombre(self):
        return self.__nombre
    def get_cedula(self):
        return self.__cedula
    def get_genero(self):
        return self.__genero
    def __str__(self):
        return f"Nombre= {self.get_nombre()} \nCédula= {self.get_cedula()} \nGénero= {self.get_genero()} \n"
class paciente(persona):
    def __init__(self, n, c, g, s):
        persona.__init__(self, n, c, g)
        self.__servicio=s
    def get_servicio(self):
        return self.__servicio
    def __str__(self):
        return f"Nombre= {self.get_nombre()} \nCédula= {self.get_cedula()} \nGénero= {self.get_genero()} \nServicio= {self.get_servicio()}"
class enfermera(persona):
    def __init__(self, n, c, g, t, r):
        persona.__init__(self, n, c, g)
        self.__turno=t
        self.__rango=r
    def get_turno(self):
        return self.__turno
    def get_rango(self):
        return self.__rango
    def __str__(self):
        return f"Nombre= {self.get_nombre()} \nCédula= {self.get_cedula()} \nGénero= {self.get_genero()} \nServicio= {self.get_turno()} \nRango= {self.get_rango()}"
class medico(enfermera):
   def __init__(self, n, c, g, t, e):
       enfermera.__init__(self, n, c, g, t, None)
       self.__especialidad=e
       del self.rango
   def get_especialidad(self):
        return self.__especialidad
   def __str__(self):
        return f"Nombre= {self.get_nombre()} \nCédula= {self.get_cedula()} \nGénero= {self.get_genero()} \nEspecialidad {self.get_especialidad()}"
def vernum(msj):
    try:
        x=int(input(msj))
    except:
        print("Ingrese un número entero.")
    return x
def vergen():
    sex=vernum("Ingrese el género: \n1. Macho. \n2. Hembra. \n")
    if sex==1:
        gen=("Macho")
    elif sex==2:
        gen=("Hembra")
    else:
        print("Ingrese una opción válida.")
    return gen
while True:
    msj=("Bienvenido al sistema de gestión \nIngrese que tipo de persona desea añadir al sistema: \n1. Paciente. \n2. Enfermera. \n3. Médico. \n4. Buscar por id. \n5. Salir. \n")
    x= vernum(msj)
    if x==1:
        name=input("Ingrese el nombre del paciente: \n")
        id=vernum("Ingrese el numero de cedula: \n")
        gen = vergen()
        servicio=input("Ingrese donde se encuentra alojado. \n")
        pac= paciente(name,id,gen,servicio)
        BD[id]=pac
    elif x==2:
        name=input("Ingrese el nombre del enfermero: \n")
        id=vernum("Ingrese el numero de cedula: \n")
        gen = vergen()
        turn=input("Ingrese la hora a la que empieza y en la que finaliza (inicio-fin): \n")
        range=input("Ingrese el rango: \n")
        enf= enfermera(name,id,gen,turn,range)
        BD[id]=enf
    elif x==3:
        name=input("Ingrese el nombre del medico: \n")
        id=vernum("Ingrese el numero de cedula: \n")
        gen = vergen()
        turn=input("Ingrese la hora a la que empieza y en la que finaliza (inicio-fin): \n")
        espec=input("Ingrese la especialidad: \n")
        med= enfermera(name,id,gen,turn,espec)
        BD[id]=med
    elif x==4:
        informacion=vernum("Ingrese el ID de la persona \n")
        if informacion in BD:
            print(BD[informacion])
    elif x==5:
        print("Adiós")
        break
    else:
        print("Ingrese una opción válida")
        continue
```
